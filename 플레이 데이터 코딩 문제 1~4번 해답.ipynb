```python
n = int(input())
arr = [list(map(int, input().split())) for _ in range(n)]

l, r, u, d = n, 0, n, 0

for i in range(n):
    for j in range(n):
        if arr[i][j]:
            l = min(l, j)
            u = min(u, i)
            r = max(r, j)
            d = max(d, i)


print((r - l + 1) * (d - u + 1))
```

```python
n = int(input())

dic = {1:[1,1], 2:[2,2]}

for i in range(3, n+1):
	temp = dic[i-1]
	dic[i] = [temp[0] + temp[1], temp[0] + 1]
	
print(dic[n][0] % 100000007, dic[n][1] % 100000007)
```

```python
n = int(input())
dp = list()
dp.append(1)
dp.append(2)
dp.append(3)
for i in range(3, n+1):
    if i % 2 == 0:
        dp.append(dp[i-1] + i)
    else:
        dp.append(dp[i-3])

print(dp[n])
```
