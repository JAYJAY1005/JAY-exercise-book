** 거리계산

- 리스트를 이용하여 점 사이의 거리를 계산 해 보자.
- 직교 좌표 위에서 A는 (1, 1), B는 (3, 2), C는 (5,7)일 때 X(2, 3)와 A/B/C와의 거리를 각각 구하여라.
- 위 코드블럭을 이용하여 distMeasure 함수를 정의하라.

- 파이썬 이용한 버전

```python
import math

class distMeasure:
    def __init__(self, x,y):
        self.x = x
        self.y = y
        
A = distMeasure(x=1, y=1)    
B = distMeasure(x=3, y=2)    
C = distMeasure(x=5, y=7)
D = distMeasure(x=2, y=3)    

print('A: {} {}'.format(A.x, A.y))    
print('B: {} {}'.format(B.x, B.y))
print('C: {} {}'.format(C.x, C.y))
print('D: {} {}'.format(D.x, D.y))
```

```python
#  거리 계산
E = D.x - A.x    
F = D.y - A.y

G = B.x - D.x
H = D.y - B.y

I = C.x - D.x
J = C.y - D.y

resultA = math.sqrt((E*E)+(F*F))
resultB = math.sqrt((G*G)+(H*H))
resultC = math.sqrt((I*I)+(J*J))
```

```python
print(resultA,resultB,resultC)
```

- 넘파이 이용한 버전

```python


import numpy as np
a = np.array((1,1))
b = np.array((3,2))
c = np.array((5,7))
x = np.array((2,3))

resultA = np.linalg.norm(x-a) # norm 함수를 이용하면 거리 값을 쉽게 가져올수 있음
resultB = np.linalg.norm(x-b)
resultC = np.linalg.norm(x-c)

print(resultA,resultB,resultC)
```
